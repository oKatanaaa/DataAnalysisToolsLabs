```python
# 2.8

def generate_noise(dataset, p=0.5):
    # Случайным образом вставляет nan в таблицу
    # p - вероятность вставки nan в строку
    dataset = dataset.copy()
    for i in range(dataset.shape[0]):
        if np.random.random() < p:
            continue
        column = np.random.choice(dataset.columns[:-1])  # avoid selecting class column
        dataset.loc[i, column] = np.nan
    return dataset

test_ds = generate_noise(dataset[:5])
test_ds
```

```python
# Выбросим строки с Nan
test_ds.dropna()
```

```python
# Заменим Nan на средние значения
column_means = test_ds.mean(axis=0)
for col in test_ds.columns:
    test_ds[col].fillna(column_means[col], inplace=True)
test_ds
```

Найдём среднее число спальных комнат при заданном возрасте дома и значении MedInc:

```python
# Сделаем некоторое округление, чтобы всё стало красивее
def round_fn(row):
    row['MedInc'] = int(row['MedInc'])
    return row
```

```python
dataset_cp = list(map(round_fn, [row for i, row in dataset.iterrows()]))
dataset_cp = pd.DataFrame.from_records(dataset_cp)
```

```python
dataset_cp.pivot_table(columns='MedInc', index='HouseAge', values='AveBedrms', )
```

Практически везде в среднем по одной спальной комнате, но в более молодых домах и домах с небольшим значением MedInc имеют большее число спален.
